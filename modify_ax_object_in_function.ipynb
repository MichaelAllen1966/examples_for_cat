```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(10)
y = x ** 3
```

```python
def modify_ax_object(ax, x, y):
    """
    Build an ax object after being passed an empty ax object.
    
    """
    
    ax.plot(x,y)
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('X Cubed')
    ax.grid()
    
    return ax    
```

```python
fig = plt.figure()
ax = fig.add_subplot()
ax = modify_ax_object(ax, x, y)
plt.show()
```
