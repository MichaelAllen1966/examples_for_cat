```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(10)
y = x ** 3
```

```python
def make_ax_object(ax, x, y):
    ax.plot(x,y)
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    return ax    
```

```python
fig = plt.figure()
ax = fig.add_subplot()
ax = make_ax_object(ax, x, y)
plt.show()
```
